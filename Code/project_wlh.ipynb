```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
quote_page = 'https://baike.baidu.com/item/%E7%8E%8B%E5%8A%9B%E5%AE%8F'
#quote_page ='https://baike.so.com/doc/2749961-2902256.html#2749961-2902256-3_3'
def get_html(url):
 #Get the html as soup
    s = requests.Session()
    s.headers['User-Agent'] = 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_9_2) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/34.0.1847.131 Safari/537.36'
    page = s.get(quote_page)
    page.encoding = 'utf-8'
    html = BeautifulSoup(page.text, 'html.parser')
    return html
```

```python
Name=[]

Date=[]
Location=[]

html = get_html(quote_page)

anchor_card = html.find('table', attrs={'log-set-param': 'table_view', 'class': 'horizontal-module j-module-5223342 j-common-module'})
concert_summary = anchor_card.find_all('td', class_='normal ')

for k, concert in enumerate(list(anchor_card.find_all('td', class_='submodule'))):
    print('-----------------------', concert_summary[3 * k].get_text().strip(), concert_summary[3 * k + 1].get_text().strip(), concert_summary[3 * k + 2].get_text().strip(), '-----------------------')
    rounds = concert.find_all('span', class_="column")
    for r in range(0, len(rounds), 3):
        print('    ', rounds[r + 1].get_text().strip(), rounds[r + 2].get_text().strip())
        Name.append(str(rounds[r + 1].get_text().strip())+','+str(rounds[r + 2].get_text().strip()))    
```

```python
print(len(Name))
```

```python
for i in Name:
    i=i.split(',')
    #print(i[0])
    Date.append(i[0])
    Location.append(i[1])
```

```python
df = pd.DataFrame({'date':Date,
                   'location':Location})
df
```

```python
df.to_csv('wanglihong_anchor_rev20181128.csv', encoding='utf-8')
```
